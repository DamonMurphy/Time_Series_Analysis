**Import any necessary libraries.**

```python
import numpy as np
import pandas as pd
import datetime as datetime
```

**Read in the data UMTMVS.csv file from the Data folder**

**Check the head of the data**

```python
# CODE HERE
df.set_index('DATE',inplace=True)
df.head()
```

**Check the data type of the index.**

```python
# CODE HERE
df.index
```

```python
df.index = pd.to_datetime(df.index)
df.index
```

**What was the percent increase in value from Jan 2009 to Jan 2019?**

```python
#CODE HERE
((df['UMTMVS'].loc['2019-01-01'] / df['UMTMVS'].loc['2009-01-01']) - 1) * 100
```

```python
((df['UMTMVS'].loc['2009-01-01'] / df['UMTMVS'].loc['2008-01-01']) - 1) * 100
```

```python
#CODE HERE
((df['UMTMVS'].loc['2009-01-01'] / df['UMTMVS'].loc['2008-01-01']) - 1) * 100
```

```python
# CODE HERE
#df['YearMean'] = df['UMTMVS'].resample('A').mean()
#df['UMTMVS'].resample('A').mean()

df.sort_values(by='UMTMVS',ascending=False).head(6)
```

**How many millions of dollars in value was lost in 2008? (Another way of posing this question is what was the value difference between Jan 2008 and Jan 2009)**

```python
# CODE HERE
df['UMTMVS'].loc['2009-01-01'] - df['UMTMVS'].loc['2008-01-01']
```

**Create a bar plot showing the average value in millions of dollars per year**

```python
# CODE HERE
df['UMTMVS'].resample('A').mean().plot.bar(color='blue',figsize=(15,8))
#df['UMTMVS'].resample('Y').mean().plot.bar(figsize=(15,8),color='blue')
```

**What year had the biggest increase in mean value from the previous year's mean value? (Lots of ways to get this answer!)**

[HINT for a useful method](https://pandas.pydata.org/pandas-docs/version/0.21/generated/pandas.DataFrame.idxmax.html)

```python
# CODE HERE
#df['UMTMVS'].resample('A').mean().sort_values(ascending=True).head()
yearly_data = df.resample('Y').mean()
yearly_data_shift = yearly_data.shift(1)
change = yearly_data - yearly_data_shift
change.idxmax()
```

```python
df['UMTMVS'].resample('A').mean().idxmax()
```

**Plot out the yearly rolling mean on top of the original data. Recall that this is monthly data and there are 12 months in a year!**

```python
# CODE HERE
#df['UMTMVS'].resample('A').mean().rolling(window=12)
df['UMTMVS_12MA'] = df['UMTMVS'].rolling(window=12).mean()
#['UMTMVS','UMTMVS_12MA']
df.plot(y=['UMTMVS','UMTMVS_12MA'],figsize=(12,5)).autoscale(enable=True,axis='x',tight=True)
```

**BONUS QUESTION (HARD).**

**Some month in 2008 the value peaked for that year. How many months did it take to surpass that 2008 peak? (Since it crashed immediately after this peak) There are many ways to get this answer. NOTE: I get 70 months as my answer, you may get 69 or 68, depending on whether or not you count the start and end months. Refer to the video solutions for full explanation on this.**

```python
#CODE HERE
```

```python
#val1 = df
max2008 = df.loc['2008-01-01':'2008-12-31','UMTMVS'].max()
month2008 = df.loc['2008-01-01':'2008-12-31','UMTMVS'].idxmax()
df2 = df.loc['2009-01-01':,:]
df2 = df2[df2['UMTMVS']>max2008]
print(month2008)
print(df2.index[0])
print(df2.index[0] - month2008)
print(((df2.index[0] - month2008)/np.timedelta64(1, 'M')))
print(df.index.get_loc(month2008))
print(df.index.get_loc(df2.index[0]))
print(df.index.get_loc(df2.index[0]) - df.index.get_loc(month2008))
```
